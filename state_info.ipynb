```python
state_codes = ['MO', 'NC', 'WI', 'NY', 'WV', 'LA', 'TX', 'NJ', 'FL', 'TN', 'GA',
       'WA', 'AR', 'IL', 'CA', 'WY', 'MN', 'CO', 'DC', 'PA', 'ME', 'IN',
       'MI', 'OR', 'OK', 'UT', 'AL', 'NV', 'AZ', 'CT', 'OH', 'VA', 'HI',
       'NM', 'MD', 'MA', 'KY', 'NH', 'ND', 'MS', 'DE', 'SC', 'KS', 'NE',
       'AK', 'RI', 'VT', 'MT', 'ID', 'SD']

state_names = dict.fromkeys(state_codes)
for key in state_codes:
    state_names[key] = state_hash[key]
state_names['DC'] = 'District of Columbia'
print(state_names)
```

```python
links_list = dict.fromkeys(state_codes)
for state in state_codes:
    links_list[state] = ('https://www.census.gov/quickfacts/fact/csv/' + state.lower() + ',US/PST045217')
```

```python
links_list
```

```python
mock_data = pd.read_csv('https://www.census.gov/quickfacts/fact/csv/dc,US/PST045217', encoding = 'utf-8')
states_data = pd.DataFrame()
states_data = states_data.append(mock_data['Fact'])
states_data
```

```python
for state in links_list:
    data = pd.read_csv(links_list[state])
    states_data = states_data.append(data[state_names[state]])
```

```python
states_data.columns = states_data.iloc[0]
states_data = states_data.drop(states_data.index[0])
states_data.head()
```

```python
states_data = states_data.iloc[:,0:59]
states_data.head()
```

```python
states_data['state_name'] = states_data.index
```

```python
states_data['addr_state'] = ''
for abb, name in state_names.items():
    for index, row in states_data.iterrows():
        if index == name:
            row['addr_state'] = abb
            
states_data.head()
```
